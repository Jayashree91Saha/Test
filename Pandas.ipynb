# Creating DataFrames in different ways

```python
#using numpy array
array1 = np.random.randint(6,10,(3,4))
print(array1)
df1 = pd.DataFrame(array1, columns = ['Rani','Rima','Sunny','Raj'], index = ['a', 'b', 'c'])
df1
```

```python
#using list of list
my_list = [['A', 25, 45, 100], ['B', 26, 56, 500], ['C',38, 67, 900]]
print(my_list)
df2 = pd.DataFrame(my_list, columns = ['Name', 'Age', 'Weight', 'Salary'])
df2
```

```python
#using lists and Zip function
name = ['Rim', 'Tom', 'Sam']
age = [23, 34, 25]
my_list = list(zip(name, age))
print(my_list)
df3 = pd.DataFrame(my_list, columns = ['Name', 'Age'] )
df3
```

```python
#using dictionary
#using list of dictionary
dict1 = {'Name':['Ani', 'Nim', 'Rim'], 'Age':[21, 28, 34], 'Weight': [56, 48, 76]}
print(dict1)
df4 = pd.DataFrame(dict1)
df4
```

```python
dict2 = [{'Name': 'Ani','Age': 21,'Weight': 56}, {'Name': 'Nim', 'Age':28, 'Weight': 48}]
print(dict2)
df5 = pd.DataFrame(dict2)
df5
```
